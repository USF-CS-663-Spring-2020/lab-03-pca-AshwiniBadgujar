```python
import pandas as pd
```

```python
df = pd.read_csv("gisette_valid.data", sep=" ",header=None);
df
```

```python
feature_df = df.iloc[:,:5000]
feature_df.describe()
```

```python
from sklearn.preprocessing import StandardScaler

x = feature_df.values
x = StandardScaler().fit_transform(x)
```

```python
import numpy as np
np.mean(x), np.std(x)
```

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
principal_breast_df = pd.DataFrame(data=principalComponents_breast, \
                                  columns = ['principal component 1', 'principal component 2'])
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.figure()
plt.figure(figsize=(10, 15))
plt.xlabel('PC 1')
plt.ylabel('PC 2')
print(principal_breast_df)
plt.scatter(principal_breast_df['principal component 1'],
            principal_breast_df['principal component 2'])
```

```python
pca = PCA().fit(x)

evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.6f}'.format(*k) for k in enumerate(evr)))
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance');
```
